## Standardization


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df=pd.read_csv("Social_Network_Ads.csv")
df.head()
```

```python
df2=df.iloc[:,2:]
```

```python
df2
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(df2.iloc[:,:2], df2.iloc[:,-1], test_size=0.2)
```

```python
X_train.head()
```

```python
X_test.head() 
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()

ss.fit(X_train)
```

```python
X_train_ss=ss.transform(X_train)
X_test_ss=ss.transform(X_test)
```

```python
X_train_ss
```

```python
X_train.describe()
```

```python
X_train_ss=pd.DataFrame(X_train_ss, columns=X_train.columns)
```

```python
X_train_ss
```

```python
np.round(X_train_ss.describe(), 1)
```

### Effect of Standardization 

### Scatter

```python
fig,(ax1, ax2)= plt.subplots(ncols=2)

ax1.scatter(X_train["Age"], X_train["EstimatedSalary"])
ax1.set_title("Before Scaling")
ax2.scatter(X_train_ss["Age"], X_train_ss["EstimatedSalary"])
ax2.set_title("After Scaling")
```

### Kde

```python
fig,(ax1, ax2)= plt.subplots(ncols=2)

ax1.set_title("Before Scaling")
sns.kdeplot(X_train["Age"], ax=ax1)
sns.kdeplot(X_train["EstimatedSalary"], ax=ax1)

ax2.set_title("After Scaling")
sns.kdeplot(X_train_ss["Age"], ax=ax2)
sns.kdeplot(X_train_ss["EstimatedSalary"], ax=ax2)
```

### Comparsion of Distm

```python
fig,(ax1, ax2)= plt.subplots(ncols=2)

ax1.set_title("Age dist before scaling")
sns.kdeplot(X_train["Age"], ax=ax1)

ax2.set_title("Age dist after Scaling")
sns.kdeplot(X_train_ss["Age"], ax=ax2)
```

```python
fig,(ax1, ax2)= plt.subplots(ncols=2)

ax1.set_title("EstimatedSalary dist before scaling")
sns.kdeplot(X_train["EstimatedSalary"], ax=ax1)

ax2.set_title("EstimatedSalary dist after Scaling")
sns.kdeplot(X_train_ss["EstimatedSalary"], ax=ax2)
```

```python
sns.scatterplot(x=X_train["Age"], y=X_train["EstimatedSalary"])
```

```python
sns.scatterplot(x=X_train_ss["Age"], y=X_train_ss["EstimatedSalary"])
```

#### Normalization Scaling

```python
from sklearn.preprocessing import MinMaxScaler
mm= MinMaxScaler()
X_train_mm=pd.DataFrame(mm.fit_transform(X_train), columns=X_train.columns)
X_test_mm=pd.DataFrame(mm.transform(X_test), columns=X_test.columns)
```

```python
X_train_mm.shape
```

```python
X_test_mm.shape
```

```python
X_train_mm.describe()
```

### Effect of Normalization

##### Scatter

```python
fig,(ax1, ax2)= plt.subplots(ncols=2)

ax1.scatter(X_train["Age"], X_train["EstimatedSalary"])
ax1.set_title("Before Scaling")
ax2.scatter(X_train_mm["Age"], X_train_mm["EstimatedSalary"])
ax2.set_title("After Scaling")
```

#### Kde

```python
fig,(ax1, ax2)= plt.subplots(ncols=2)

ax1.set_title("Before Scaling")
sns.kdeplot(X_train["Age"], ax=ax1)
sns.kdeplot(X_train["EstimatedSalary"], ax=ax1)



ax2.set_title("After Scaling")
sns.kdeplot(X_train_mm["Age"], ax=ax2)
sns.kdeplot(X_train_mm["EstimatedSalary"], ax=ax2)
```

### Comparsion of Dist

```python

fig,(ax1, ax2)= plt.subplots(ncols=2)

ax1.set_title("Age dist before scaling")
sns.kdeplot(X_train["Age"], ax=ax1)

ax2.set_title("Age dist after Scaling")
sns.kdeplot(X_train_mm["Age"], ax=ax2)
```

```python
fig,(ax1, ax2)= plt.subplots(ncols=2)

ax1.set_title("EstimatedSalary dist before scaling")
sns.kdeplot(X_train["EstimatedSalary"], ax=ax1)

ax2.set_title("EstimatedSalary dist after Scaling")
sns.kdeplot(X_train_mm["EstimatedSalary"], ax=ax2)
```

## Maxabs Scaler

Max absolute scaler is useful if there is spare data(lots of 0's in data)

```python
from sklearn.preprocessing import MaxAbsScaler
ma=MaxAbsScaler()

X_train_ma=ma.fit_transform(X_train)
X_test_ma=ma.transform(X_test)
```

```python
X_train_ma.shape, X_test_ma.shape
```

## Robust Scaler

Robust Scaling is use full if there is outliers in data

```python
from sklearn.preprocessing import RobustScaler
rs= RobustScaler()
X_train_rs=rs.fit_transform(X_train)
X_test_rs=rs
```
